```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

url = "http://www.basketball-reference.com/draft/NBA_2014.html"
html = urlopen(url)
bsobj = BeautifulSoup(html,"lxml")
names = bsobj.findAll("tr", limit = 2)

columns = names[1].findAll("th")
column_header = []
for column in columns:
    column_header.append(column.getText())
print (column_header)
    
data_rows = bsobj.findAll("tr")[2:]
player_data = []
k = 1
for i in data_rows:
    player_row = []
    player_row.append(k)
    k = k+1
    data = i.findAll("td")
    for d in data:
        player_row.append(d.getText())
    
    player_data.append(player_row)
print (player_data[1])
df = pd.DataFrame(player_data, columns=column_header)
print (df)

df.to_csv('nbaplayer.csv')
```
